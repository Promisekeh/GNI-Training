## Data Structure -Types to store Collection of data

Data structures are used to store multiple items in a single variable

* **List**
    - They are created using square brackets `[]`. items are ordered, changeable, and allow duplicate values.
* **Tuple** 
    - There are created with curved/round brackets `()`, are immutable i.e. cannot be changed. items are ordered, unchangeable, and allow duplicate values.
* **Sets**
    - They are created using curly brackets `{}`, are unordered/unindexed and unchangeable. Once a set is created, you cannot change its items, but you can add new items. Duplicates not allowed
* **Dictionaries**
    - They are created using curly brackets `{}`,  are used to store multiple items in key:value pair
    - They are ordered, changeable but don't allow duplicate keys

````python
fruits = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
my_tuple = (1, 2, 4, 5, 7)
my_set = {'james', 'john', 'mathew', 'luke'}

my_dict = {'First_Name': 'Promise',
          'Surname': 'Ekeh',
          'Nationality': 'Nigerian'}

print(f'This is a list {fruits, type(fruits)}')
print(f'This is a tuple {my_tuple, type(my_tuple)}')
print(f'This is a set {my_set,  type(my_set)}')
print(f'This is a dictionary {z, type(z), w, type(w)}')
````

# Boxplot

![](media/boxplots.jpg)

- **Median** (Q2/50th percentile): The line splitting the box in two represents the median value. This shows that 50% of the data lies on the left hand side of the median value and 50% lies on the right hand side
- **First Quartile** (Q1/25th percentile): The left edge of the box represents the first/lower quartile. it shows the value at which the first 25% of the data falls up to
- **Third Quartile** (Q3/75th percentile): The right edge of the box represents the third/upper quartile. it shows that 25% of the data lies to the right of the upper quartile value. 


- **Interquartile** Range (IQR): 25th to the 75th percentile
- **Whiskers** (shown in blue)
- **Outliers** single points (shown as green circles)
- **“Minimum”**: $Q1 - 1.5*IQR$
- “Maximum”: $Q3 + 1.5*IQR$

````python

data = pd.DataFrame({
    'category': np.repeat(['A', 'B'], 10),
    'values': np.random.randn(20)
})

# Box plot
sns.boxplot(x='category', y='values', data=data)
plt.title('Box Plot')
plt.show()
````

# Geopandas

Geopandas is an open-source project that makes working with geospatial data in Python easier. It extends Pandas to allow spatial operations on geometric types.

Key Features:

- Handling of geometric operations (intersections, unions, etc.).
- Plotting maps.
- Spatial joins and spatial indexing.

````python
# ! pip install geopandas #to install geopandas
````

````python
import geopandas as gpd


# Load a sample dataset
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plotting the world map
world.plot()
plt.title('World Map')
plt.show()
````

````python
world
````

````python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Calculate the area of each country in square kilometers
world['area_sqkm'] = world['geometry'].to_crs(epsg=3395).area / 10**6


# Plot the choropleth map
fig, ax = plt.subplots(1, 1, figsize=(20, 12))
world.boundary.plot(ax=ax)
world.plot(column='pop_est', ax=ax, legend=True,
           legend_kwds={'label': "Population Estimate"},
           cmap='OrRd')

plt.title('Choropleth Map of Population Estimate')
plt.show()
````
