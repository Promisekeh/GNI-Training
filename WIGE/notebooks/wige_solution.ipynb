## Exercise 6
`Analysing data from text file`

Create a program to analyze survey data from a text file. The data includes rock type and sample location. The program should count occurrences of each rock type.

```python
rock_data = """
Granite, Site A
Basalt, Site B
Limestone, Site A
Granite, Site B
Basalt, Site C
"""
lines = rock_data.strip().split("\n")
rock_counts = {}
for line in lines:
    rock, site = line.split(", ")
    if rock in rock_counts:
        rock_counts[rock] += 1
    else:
        rock_counts[rock] = 1
print(rock_counts)
```
